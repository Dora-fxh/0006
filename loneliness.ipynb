## 1.Introduction
The loneliness of old people has been an emerging social issue, with the ageing tendency of the population nowadays. The lonely elders always have poor health (both physical and mental) and suffer from social exclusion (Scharf and Keating, 2012; Newall, Chipperfield and Bailis, 2014). Specifically, Age UK (2016) found that people aged 65 or over were twice as likely to develop Alzheimer's disease and were probably more common to be reported symptoms of anxiety and depression if they have high levels of loneliness. Its prevalence in the UK even prompted a national campaign, which aims to reduce the negative effects of loneliness for older people (‘Campaign to end loneliness’, 2011). Most previous studies explore the loneliness of individual old people, which focus on personal information and only cover limited areas. However, our study conducts the experiment based on the community-level. It not only can cover nearly all the areas in London, but provide relatively precise estimate of the loneliness of the old and do not involve privacy issues, but how can the risk of loneliness at lower-level geographies be measured?

According to  (Rossall, Iparraguirre and Davidson, 2015) the English Longitudinal Study of Ageing (ELSA) surveyed 10,274 people aged 50+ in England and more detailed information were collected from those who answered ‘often’ or ’always’ to the question “How often do you feel lonely”. Then, regression analysis was conducted to identify all the significant factors associated with their loneliness answer. Since the sample size is still not big enough to cover all neighbourhoods, the formula derived from the national (ELSA) analysis was applied utilising census (2011) data to the small areas (LSOA) to find the loneliness indicator: ‘log odds’ of the risk of loneliness among older people. Although they are predicted values, their reliability is properly validated by many academies and they are even accepted as politicians. Thus, we can treat it as a reliable measurement of loneliness for old people. 

## 2.Literature Review
Loneliness refers to the disparity between the actual and the desired or expected levels of social contact. It is a subjective feeling about the “perceived quality of the person’s relationships”.  Several factors could contribute to it, including the health status, economic situation, transport accessibility, neighbourhood and family ties, ethnic-cultural backgrounds, education level, neighbourhood safety, and religion etc.
Neighbourhood safety and health status are significant factors to affect the loneliness level of people especially at older ages (Kearns et al., 2015; Kemperman et al., 2019). Logically, older people with a safer living environment and healthier body would feel less depressed and be more active to get in touch with others.
The economic situation may be another impact element. Some experts found the loneliness of the older in a community would be affected by the economy, and whether reducing the loneliness level for old people is cost-effective depends on the value that society places on loneliness reduction (Mihalopoulos et al., 2019; McDaid and Park, 2021). Also, the estate of the old people may affect their attitude and emotion to life (Bergefurt et al., 2019).
Family and neighbourhood bond is one of the strongest predictors of the loneliness of old. Many studies found a similar relationship between the loneliness of old people and the marital status, which shows that widowed senior citizens have the highest level of loneliness, the divorced or separated older comes second, and the married old people have the lowest loneliness level(Victor et al., 2005; Hazer and Boylu, 2010; Nzabona, Ntozi and Rutaremwa, 2016).  Dahlberg and Mckee (2014) also found that the level of contact with family and friends is positively correlated to social loneliness. 
The religious community may affect the loneliness level of the elderly. According to a survey conducted by Escher et al. (2019) on 102 adults over the age of 55, outness to the religious community is positively related to depression but negatively related to loneliness. On the contrary, a more general study, which based on 8504 older people samples, concluded that lower religious attendance may cause higher depression and loneliness level (Curran et al., 2019). A study based on Iranians and Turks found that religion is a strong tool to protect older from loneliness in the United States (Lewin and Thomas, 2001), and it is also helpful for older Romanian immigrants in the US (Ciobanu and Fokkema, 2017).  
Education level may have a sophisticated influence on the loneliness level of the old. A study based on the Chinese context found that, though literate older women have a lower loneliness level than their counterparts when they remain married, they would perform no difference to others when their spouse passed away (Yang, 2020). Another research argued that education level may affect people’s economic status, knowledge, experiences and assets, which would affect the loneliness level of old people (Jana et al. 2015).
Except for the factors mentioned above, the elderly with different ethnic-cultural backgrounds may have different average loneliness levels (De Jong Gierveld, Van der Pas and Keating, 2015).  Moreover, the distance from public spaces is also found to affect loneliness (Maas et al., 2009).

We define a function which helps us to choose the specific columns we want in a dataframe which have three level of head.

```python
def get_ratio(first_level_head, columns, dataframe=census_data, delete_level3=True):
    """ params:
        dataframe:  a dataframe with 3 levels of head
        first_head: the column name you want to choose in the first level of head 
        columns: a number or a list which is the columns you want in the "first_head"
        description:From the first head choose the column name and then choose its nth column        
    """
    ratio_df = dataframe[first_level_head]  
    if delete_level3:
        # we don't need the year-index in census_data
        ratio_df.columns = ratio_df.columns.droplevel(1)
    ratio_df = ratio_df.iloc[:,columns]
    return ratio_df
```

Then, we can use this function to get the columns we want and concat those which are from the same dataframe to new dataframes. Codes and names below are used for merging.

```python
codes = census_data["Unnamed: 0_level_0","Unnamed: 0_level_1","Codes"]
codes.name="codes"
names = census_data["Unnamed: 1_level_0","Unnamed: 1_level_1","Names"]
names.name="names" 

census_data_df = pd.concat([codes,names,get_ratio("Ethnic Group",6), get_ratio("Religion",16),\
                     get_ratio("Tenure",4), get_ratio("Qualifications",13,census_data_1)],axis=1)

codes = exclusion_df["LSOA Code","Unnamed: 0_level_1","Unnamed: 0_level_2"]
codes.name="codes"
names = exclusion_df["LSOA Name","Unnamed: 1_level_1","Unnamed: 1_level_2"]
names.name="names"

safety = exclusion_df["Safety","Fear of crime"]
exclusion_data_df = pd.concat([codes,names,get_ratio("Economic Situation",0,exclusion_df),\
                               get_ratio("Transport Accessibility",[0,1],exclusion_df),get_ratio("Household Ties",1,exclusion_df),\
                               get_ratio("Neighbourhood Ties",[0,1],exclusion_df),get_ratio("Health",0,exclusion_df),\
                               safety],axis=1)
census_data_df.head()
```

Collinearity indicates the strong correlation between regressors and therefore contain similar information about the variance within the dataset. We should eliminate it as it can cause many problems:
- The results of predictors can be wildly different since a little changes in samples can lead to hugely different effects.
- We can not make interpretation of a coefficient as the effect of a a unit increment in that variable when keep other variables unchanged.
Heatmap conducted from the correlation matrix is useful to identify it.

```python
# This function is amended from: https://stackoverflow.com/a/51329496/4667568


def drop_column_using_vif_(df, list_var_not_to_remove, thresh=5):
    '''
    Calculates VIF each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param list_var_not_to_remove: the list of variables that should not be removed even though it has a high VIF. For example, dummy (or indicator) variables represent a categorical variable with three or more categories.
    :param thresh: the max VIF value before the feature is removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constatnt item to the data
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()

        # drop the const
        vif_df = vif_df.drop('const').drop(list_var_not_to_remove)
        
        print('Max VIF:', vif_df.VIF.max())
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

### 5.1 Multiple linear regression analysis
Multiple linear regression can estimates predictor jointly by the optimal combination of two or more regressors.
 $$ y = \beta_0 + \beta_1 x_1 + ... + \beta_n x_n +\epsilon  $$ 

Here,
* y is the dependent variable and it's what we are trying to predict. 
* $\beta_0 $ is the intercept, it is a constant and when x is 0, it represents the value of y 
* $\beta_i $ is the Coefficient(or the Weight) of x_i. It means a unit increases in x_i will lead to a unit increase in y when other variables remain constant. 
* $x_i$ is the independent variable (regressor)
* $ \epsilon$ is the redidual which should be conform to the law of normally distributed and be independent ~ N( 0, 1)
$$ \text{residual}_i = y_i - \hat{y}_i $$



$R^2$ **(Coefficient of the determination)** measures the percentage of variance of the responde variable explained by the model. It's a value between 0 and 1. The closer it close to 1, the better the performance of the model.
$$ \operatorname{R^2} = \frac{SSR}{SST} $$
Sum of the Squared Regression(SSR) is the residual that can be explained by the regression line. 
 $$ SSR = \sum_{i=1}^n \left(\hat{y_i} - \bar{y}\right)^2 $$
Total Sum of the Squared Error(SST) is the total residual. 
 $$SST = SSR + SSE$$
 $$ SSE = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$



**Adjusted R-Squared($R^2_{adj}$)**: 

$R^2_{adj}$ is similar to $R^2$, but the value of $R^2_{adj}$ is usually used to compare model with different number of variables. As the following equation shows, the increase of number of feature (k) will penilizes the adjusted $R^2$ if the increased variable can not significantly improve the $R^2$.  
$$ {R^2_{adj}} = 1 - [\frac{(1 - R^2)(n-1)}{(n-k-1)}]$$

here, 
* n = number of samples. 
* k = number of variable. 

Since we have introduced the $R^2$, we can add some mathematic explanation of <b>VIF<b/>.
$$ VIF_i=\frac{1}{1-R^2_i} $$
$R^2_i$ is the $R^2$ got by regressing the variable $X_i$ on the remaining regressors

#### Residual examination
**Residuals should be independent and identically distributed. Moreover, we can examine abnormal points from following metrics**
- Leverage: A data point has high leverage if it has "extreme" predictor x values. With a single predictor, an extreme x value is simply one that is particularly high or low.
- Studentized residual depends not only on the ordinary residual, but also the size of the mean square error (MSE) and the leverage. An observation with an internally studentized residual that is larger than 3 (in absolute value) is generally deemed an outlier.
- Cook’s distance is a measure to determine the influential point. Data points that have high leverage and are outliers incline to be influential.

#### 5.2.2 Random Forest

Random forest is a popular machine learning method that can be used in both classification and regression problems to build predictive models. The key of random forest is to create a large number of A large number of relatively uncorrelated decision trees to operate as an ensemble and get the best possible answer. It always can get a better result than decision trees, since it conduct multiple learning models which protect each other from their individual errors.

Decision Tree always tend to overfit on training data since it uisng the same data to utilize the training and test sets every time, which is not robust and will lead to poor performance on unseen data. 

In order to reduce the correlation among the individual trees, random forest chooses only a subsample from the feature space at each split. And it always has shallow tree structure for each individual trees inside it to make the trees uncorrelated and avoid overfitting. Thus, it has several method to set a stopping criteria for node splits, like the maximum depth of the tree, the minimal impurity split threshold and so on.
**Parameter selection**

We will tune three main hyperparameters that may effect the efficiency of our random forest model:

* max_features: The maximun number of features to consider when looking for the best split. The bigger this hyperparameter is, the lower speed of the model will be. And increasing it can improve the accuracy of the individual trees, but will also reduces the diversity of the whole forest. 
* n_estimators: The number of individual decision trees of the random forest. More individual trees can improve the efficiency but will reduce the speed of the model.
* min_samples_leaf: The minimum number of samples required to be at a leaf node. This may have the effect of smoothing the model if we have more than one sample at each leaf node, especially in regression.

Though there are many other hyperparameters to tune, they have the similar effect on the model as the three above.

**K-fold cross validation**
To avoid over-learning or under-learning, and make the results more persuasive, K-fold cross-validation (K-CV) is applied to select our parameters. The training data is randomly divided into K groups (here is 5) on average. Each subset of data would be used as validation set once, and the remaining K-1 subset data is used as the new training set, which means K-CV enables each sample data to be used as both training and validation data. Therefore, K models are obtained in total, and the average $ R^2 $ of these K models’ validation set is used as the performance index under K-CV(since the default criterion of random forest is MSE). 



```python
def cv_to_choose_parameter(parameters):
    '''
    param: parameters is a dictory, it represent the parameter and their values we want to choose
    description:use grid research to find the parameter value which has the smallest MSE and output the best parameter and the corresponding score
    '''
    rf = RandomForestRegressor()
    clf = model_selection.GridSearchCV(rf, parameters)
    clf.fit(X_train, y_train)
    means = clf.cv_results_['mean_test_score']
    params = clf.cv_results_['params']
    for mean,param in zip(means,params):
        print("Score %f with: %r" % (mean, param))
    # we can query the best parameter value and its accuracy score
    print ("The best parameter value is: %r The best score is: %f" %(clf.best_params_,clf.best_score_) )
    return clf.best_params_
   
```

#### 5.2.3 Regularization Models
From the book written by Raschka (2019), we know regularization is helpful when we need to prevent overfitting, improving generalization, handle collinearity and filter out noise from data. The key point of regularization is to introduce an additional bias item , which imposes the penalty on the coefficients.By implement "regularization", the model can reduce the complexity of the model.

There are three types of regularizations which we will experiment in the project. 
* **Ridge**
* **Lasso**
* **Elastic Net**

They will all penalize the magnitude of the coefficient, and simultaneously work as the normal LR (Linear Regression) to minimize the error between actual observed values and the predicted one. Comparing to the LR, the loss function for regularization models is to minimize $E(Y,\hat{Y})+\alpha||w||$, where $E(Y,\hat{Y})$ is the loss functino for LR, $||*||$ is typically L1 or L2 norm and $\alpha$ is a hyperparameter determining the amount of regularization. This loss function balance the error of predicting and the magnitude of the coefficients. 

The difference for the three kind of regularizations is only the way to penalize the coefficients. Ridge regression is an L2 penalized model where we minimize the squared sum of the coefficients. Lasso regression is an L1 penalized model where we minimize the absolute sum of the coefficients. Ridge method can better deal with multicollinearity and reduce the effect of low significant features. Lasso is more suited for redundant variables, since it will force the coefficients of them to be zero. Lasso yields sparse feature vectors, which is useful when we have many irrelevant features. Elastic Net combines the two and put the ratio to minimize the squared and absolute sum of the coefficients. 

To see clearly the difference among them, we have a look at the loss function of them.

##### Ridge:

 $$ \text{minimize:}\; RSS+Ridge = \sum_{i=1}^n \left(y_i - \left(\beta_0 + \sum_{j=1}^p\beta_j x_j\right)\right)^2 + \alpha\sum_{j=1}^p \beta_j^2$$

where, 
* $\alpha$ is the hyperparameter we need to define. A bigger value of $\alpha$ will have more impact on the loss function. 
    * When $\alpha$ is 0, it is just the LR. 
    * When $\alpha$ is $\infty$, all coefficients of our features would become 0
    * When $\alpha$ is between  0 and $\infty$, The $\alpha$ parameter will determine the magnitude of the coefficients. The coefficients will be between 0 and 1 for LR. 
* $\sum_{j=1}^p \beta_j^2$, is the squared sum of all the coefficients. 


##### Lasso:

$$ \text{minimize:}\; RSS + Lasso = \sum_{i=1}^n \left(y_i - \left(\beta_0 + \sum_{j=1}^p\beta_j x_j\right)\right)^2 + \alpha\sum_{j=1}^p |\beta_j|$$

Where, 
* $\alpha$ is a hyperparameter similar to the Ridge model. 
* $\sum_{j=1}^p |\beta_j|$ is the absolute sum of the coefficients.

##### Elastic Net: 

$$ \text{minimize:}\; RSS + Ridge + Lasso = \sum_{i=1}^n \left(y_i - \left(\beta_0 + \sum_{j=1}^p\beta_j x_j\right)\right)^2 + \alpha_1\sum_{j=1}^p |\beta_j| + \alpha_2\sum_{j=1}^p \beta_j^2$$
Where,
* $\alpha_1\sum_{j=1}^p |\beta_j|$ is the penality item as Lasso model.
* $\alpha_2\sum_{j=1}^p \beta_j^2$ is the penality item as Ridge model.

**ElasticNet regression**

```python
# 4* ElasticNet
elasticNetCV = ElasticNetCV(l1_ratio = [1e-4,0.001,0.01,0.05,0.1, 0.3, 0.5, 0.6, 0.7, 0.8, 0.85, 0.9, 0.95, 1],
                          alphas = [1e-7,1e-6,1e-5,5e-5, 1e-4, 5e-4, 1e-3, 5e-3, 0.01, 0.05, 0.1, 0.5, 1], 
                          max_iter = 50000, cv = 10,normalize = True)# l1_ratio is a list, in which case the different values are tested by cross-validation and the one giving the best prediction score is used.
elasticNetCV.fit(X_train, y_train)
alpha = elasticNetCV.alpha_
ratio = elasticNetCV.l1_ratio_
print("Best l1_ratio :", ratio)
print("Best alpha :", alpha )

print("Try again for more precision with l1_ratio centered around " + str(ratio))
elasticNetCV = ElasticNetCV(l1_ratio = [ratio * .5, ratio * .75, ratio * .80,ratio * .85, ratio * .9, ratio * .95, ratio, ratio * 1.05, ratio * 1.1],
                          alphas = [1e-7,5e-7,1e-6,5e-6,1e-5,5e-5, 8e-5, 1e-4, 3e-4, 3e-4,5e-4, 7e-4, 5e-3, 1e-3, 0.01], 
                          max_iter = 50000, cv = 10,normalize = True)
elasticNetCV.fit(X_train, y_train)
if (elasticNetCV.l1_ratio_ > 1):
    elasticNetCV.l1_ratio_ = 1    
alpha = elasticNetCV.alpha_
ratio = elasticNetCV.l1_ratio_
print("Best l1_ratio :", ratio)
print("Best alpha :", alpha )

print("Now try again for more precision on alpha, with l1_ratio fixed at " + str(ratio) + 
      " and alpha centered around " + str(alpha))
elasticNetCV = ElasticNetCV(l1_ratio = ratio,
                          alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, alpha * .85, alpha * .9, 
                                    alpha * .95, alpha, alpha * 1.05, alpha * 1.1, alpha * 1.15, alpha * 1.25, alpha * 1.3, 
                                    alpha * 1.35, alpha * 1.4], 
                          max_iter = 50000, cv = 10,normalize = True)
elasticNetCV.fit(X_train, y_train)
if (elasticNetCV.l1_ratio_ > 1):
    elasticNetCV.l1_ratio_ = 1    
alpha = elasticNetCV.alpha_
ratio = elasticNetCV.l1_ratio_

# using the best alpha and l1_ratio we get and the whole training and validation set to fit the model
elasticNet = ElasticNet(alpha=alpha,normalize = True)
elasticNet.fit(X_train, y_train)
y_train_ela = elasticNet.predict(X_train)
print("Best l1_ratio : %f and Best alpha : %f with R2: %f" % (ratio, alpha, elasticNet.score(X_train, y_train)))

y_train_ela = elasticNet.predict(X_train)

# Plot residuals
plt.scatter(y_train_ela, y_train_ela - y_train, c = "blue", marker = "s", label = "Training data")
plt.title("Figure 13: Linear regression with ElasticNet regularization")
plt.xlabel("Predicted values")
plt.ylabel("Residuals")
plt.legend(loc = "upper left")
plt.hlines(y = 0, xmin = -5.5, xmax = -2, color = "red")
plt.show()


# Plot important coefficients
coefs = pd.Series(elasticNet.coef_, index = X_train.columns)
print("ElasticNet picked " + str(sum(coefs != 0)) + " features and eliminated the other " +  str(sum(coefs == 0)) + " features")
coefs.plot(kind = "barh")
plt.title("Figure 14: Coefficients in the ElasticNet Model")
plt.show()
```

L1 ratio is the scaling between l1 and l2 penalties (For l1_ratio = 0 the penalty is an L2 penalty. For l1_ratio = 1 it is an L1 penalty. For 0 < l1_ratio < 1, the penalty is a combination of L1 and L2). More specifically, the optimization objective is (from scikit-learn website):
$$ 1 / (2 * nsamples) * ||y - Xw||^2_2 + alpha * l1ratio * ||w||_1
+ 0.5 * alpha * (1 - l1ratio) * ||w||^2_2 $$
It is equivalent to:
$$ \alpha_1\sum_{j=1}^p |\beta_j| + \alpha_2\sum_{j=1}^p \beta_j^2 $$
$$ alpha = \alpha_1 + \alpha_2$$ and $$l1ratio = \alpha_1 / (\alpha_1 + \alpha_2) $$
The optimal L1 ratio used by ElasticNet here is equal to 0.00005, which means it is almost equal to the Ridge regressor we used earlier. The model seems do not need many L1 regularization to overcome potential L2 shortcoming.
In the presence of high collinearity, ridge is better than Lasso
The advantage of L1 regularization is that it is more robust against outliers as it uses the absolute value between the estimated outlier and the penalization term. In this case, although the outlier still exist, its problems is not severe.

Here
* $y_i$ = Each observed data point. 
* $\hat{y_i}$ = Predicted data point for each $x_i$ depending on i.

## 6.Conclusion and Limitation
In the report, multilinear regression is first performed to explore the factors which are correlated with the loneliness for the elderly and then some more advanced models is utilised to see whether they have better prediction power on the probability of the risk of loneliness among older people at lower super output areas (LSOA) in London.

We find except the percentage of 65+ people worried about anti-social, all the variables that pass the VIF test are significant and are correlated with the loneliness of the old. Among these, the percentage of old people have no cars in the household contributes most in all models since it reflects not only the most of the transport accessibility, but also the economic situation of the elderly. 

For the three main models we trained and tested, random forest outperforms others (about 74% of the total variation can be explained by the model) though it exists over-fitting problem and lacks generalization ability. Our MLR model is relatively stable and do not have severe problems of multicollinearity and outlier, so the testing score is just slightly poorer than the regularized model. This can be verified in the parameter selection part for alphas, the parameter selected for both alpha1 and alpha2 is small, especially alpha2. Additionally, ElasticNet regression put a lot more most weight on L2 penalty. Thus, the regularized model did not improve obviously.


The Limitation mainly lies on the data: 
- data are not consistent in the year, they range from 2011 to 2015 and seems to be a little old.
- some variable reflect the situation of all ages, but we use them to reflect the older people’s status. e.g., we just assume that the higher the overall level of education, the higher the level of education of the elderly
- Our model has features from many different aspects, which makes the analysis a little messy and not detailed in every features. In the future, we can explore more on one aspect. For example, using percentage of people from different religions in the communities to analyse each religion's effect on the loneliness of the old people.
- Some features (like the percentage in borough who feel unsafe walking alone after dark) only have the data based on borough level, we simply use the borough-level data for every LSOA in that area.
- Some variables are still weakly correlated, which may have the impact on our final result. Take Random Forest for example, although collinearity may not be an issue for the precision of the final result of Random Forest. However, it will lead to incorrect conclusion of the significance that the features contribute to the result.
- We find the existence of outliers, we do not deal with them as some of our models are robust to outliers. Nevertheless, more attention should be paid to them as it may reveal some potential findings.
- Four variables is transformed using box-cox, this will influence the specific interpretation of their coefficient in MLR

For the model, random forests are relatively good at generalization, but it shows big gap between the results of training set and validation set(also test set). In our project, it may due to that we only tuned three significant hyperparameters and there are many other hyperparameters corresponding to cutting branches and reducing the depth of the trees we didnot try. What's more, the MLR performs better than our expectation, it may be that our true loneliness values are predicted using regression model.
